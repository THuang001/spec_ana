```python
def prepare(path):
    file = open(path)
    with open(path+'.csv','w') as res:
        for line in file:
            if len(line)>=100:
                line_list = ' '.join(line.split())
                res.write(line_list)
                res.write('\n')
```

```python
import pandas as pd
```

```python
path = './500/pmctrack_result1'
prepare(path)
```

```python
data = pd.read_csv(path+'.csv',sep=' ')
data.head()
```

```python
def merge(path):
    data = pd.read_csv(path+'.csv',sep=' ')
    tmp = data.iloc[range(1,len(data),2)].reset_index(drop=True)
    l = len(data)//2
    ttt=data.drop(range(1,len(data),2)).reset_index(drop=True)
    ttt['pmc5'] = tmp['pmc0'].reset_index(drop=True)[:l]
    ttt['pmc6'] = tmp['pmc1'].reset_index(drop=True)[:l]
    ttt['pmc7'] = tmp['pmc3'].reset_index(drop=True)[:l]
    ttt[:l].to_csv(path+'_merge.csv')
```

```python
path_csv = './500/pmctrack_result1'
merge(path_csv)
```

```python
data_merge = pd.read_csv('./500/pmctrack_result1_merge.csv')
data_merge.head()
```

```python
def filtered(data):
    col = ['Unnamed: 0','nsample','pid','coretype','expid','event','pmc7','pmc3']
    data.drop(col,axis = 1,inplace = True)
    data.rename(columns={'pmc6':'pmc3'},inplace=True)
    data = data[['pmc0', 'pmc1', 'pmc2', 'pmc3', 'pmc4', 'pmc5']]
    data = (data-data.mean())/data.std()
    return data
```

```python
def status(x):
    return pd.Series([x.count(),x.min(),x.idxmin,x.quantile(.25),x.median(),x.quantile(.75),
                      x.mean(),x.max(),x.idxmax,x.mad(),x.var(),x.std(),x.skew(),x.kurt()],index=['总数','最小值',
                    '最小值位置','25%分位数','中位数','75%分位数','均值','最大值','最大值位置','平均绝对偏差','方差','标准差',
                                                                                                    '偏度','峰度'])
```

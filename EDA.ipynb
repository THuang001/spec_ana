```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = open('./spec2017_data/508/pmctrack_result_copy')
with open('result.txt','w') as res:
    for line in file:
        line_list = ' '.join(line.split())
        res.write(line_list)
        res.write('\n')
```

```python
data = pd.read_csv('result.txt',sep=' ')
```
